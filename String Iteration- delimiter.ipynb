```python
#STATES Usecase

rawlis = [ "State->TN~City->Chennai" , "State->Kerala~City->Trivandrum" ]


rddlis = sc.parallelize(rawlis)
print()
print("====RDDLIS====")
print(rddlis.collect())

flatdata = rddlis.flatMap(lambda x : x.split("~"))
print()
print("====flatdata====")
print (flatdata.collect())

stateslist = flatdata.filter (lambda x : 'State' in x)
print()
print("====stateslist====")
print(stateslist.collect())

citylist = flatdata.filter(lambda x : 'City' in x )
print()
print("====citylist====")
print(citylist.collect())

states = stateslist.map(lambda x : x.replace("State->" , ""))
print()
print("====states====")
print(states.collect())

cities = citylist.map(lambda x : x.replace("City->" , ""))
print()
print("====cities====")
print(cities.collect())
```
